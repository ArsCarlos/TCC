````python
import os

# Melhorar filtragem de dados

def find_files(path):
# r=root, d=directories, f = files
    files = []
    for r, d, f in os.walk(path):
        for file in f:
            if '.dbf' in file:
                files.append(os.path.join(r, file))
    return files

path_sinasc_2017 = '../DATASUS_DADOS_1996_A_2017_DBF/SINASC/2017/'
path_sinasc_2016 = '../DATASUS_DADOS_1996_A_2017_DBF/SINASC/2016/'
path_sim = '../DATASUS_DADOS_1996_A_2017_DBF/SIM/2017'

sinasc_2017 = find_files(path_sinasc_2017)
sinasc_2016 = find_files(path_sinasc_2016)
sim_files = find_files(path_sim)

sinasc_files = sinasc_2016 + sinasc_2017
all_files = sinasc_files + sim_files

for f in all_files:
    print(f)
````

# LINKAGE

````python
# Filtrando os rows que devem ser avaliados.
missing_sim = sim_ap[sim_ap['NUMERODN'].isna()]
missing_sim["PESO"] = pd.to_numeric(missing_sim["PESO"])
new_sinasc["PESO"] = pd.to_numeric(new_sinasc["PESO"])

new_sinasc.columns
````

````python
from recordlinkage.index import Block
from recordlinkage import Compare

indexer = Block(
    left_on=['CODMUNNATU', 'DTNASC'],
    right_on=['CODMUNNATU', 'DTNASC']
)

"""
Replicando: http://www.scielo.br/pdf/rbsmi/v15n1/1519-3829-rbsmi-15-01-0057.pdf
sexo da criança: OK
idade da mãe: OK 
raça/cor: OK 
escolaridade materna: OK 
ocupação da mãe: OK
número de filhos: OK
nascidos vivos: OK
número de filhos nascidos mortos: Não
tipo de parto: Ok
duração da gestação: Ok 
peso ao nascer: Ok
tipo de gravidez: Ok
"""

#new_sinasc = new_sinasc.reset_index()
candidate_links = indexer.index(missing_sim, new_sinasc)

print(len(candidate_links))
compare = Compare()


compare.exact('PARTO', 'PARTO', label='PARTO')
compare.exact('RACACOR', 'RACACOR', label='RACACOR')
compare.exact('SEXO', 'SEXO', label='SEXO')
compare.exact('IDADEMAE', 'IDADEMAE', label='IDADEMAE')
compare.numeric('PESO', 'PESO', label='PESO')
compare.exact('SEMAGESTAC', 'SEMAGESTAC', label='SEMAGESTAC')
compare.exact('QTDFILMORT', 'QTDFILMORT', label='QTDFILMORT')
compare.exact('QTDFILVIVO', 'QTDFILVIVO', label='QTDFILVIVO')
compare.exact('ESCMAE', 'ESCMAE', label='ESCMAE')
#compare.exact('OCUPMAE', 'OCUPMAE', label='OCUPMAE')
compare.exact('GRAVIDEZ', 'GRAVIDEZ', label='GRAVIDEZ')

#compare.exact('NUMERODO', 'NUMERODN', label='NUMERO')

compare_vectors = compare.compute(candidate_links, missing_sim, new_sinasc)

compare_vectors.describe()
````

````python
class bcolors:
    HEADER = '\033[95m'
    OKBLUE = '\033[94m'
    OKGREEN = '\033[92m'
    WARNING = '\033[93m'
    FAIL = '\033[91m'
    ENDC = '\033[0m'
    BOLD = '\033[1m'
    UNDERLINE = '\033[4m'
````

````python
same_fields = df_sim.columns.isin(df_sinasc.columns)
match_fields = list(df_sim.columns[same_fields])


def compare_matches(sim_index, sinasc_index):
    for field in match_fields:
        if missing_sim.loc[sim_index][field] == new_sinasc.loc[sinasc_index][field]:
            print(bcolors.OKGREEN, "field ", field, ': ', missing_sim.loc[sim_index][field], new_sinasc.loc[sinasc_index][field], bcolors.ENDC)
        else:
            print("field ", field, ': ', missing_sim.loc[sim_index][field], new_sinasc.loc[sinasc_index][field])

compare_matches(9029, 19753)
print('---------validar:----------')
compare_matches(2196, 11849)
print('-------------------')
compare_matches(2196, 11878)
print('-------------------')
````
